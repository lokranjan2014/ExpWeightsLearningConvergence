# Exponential Weights Simulation for 2×2 Symmetric Games

We demonstrate Algorithm 1 (Exponential Weights) for 2×2 symmetric games, computing the shorthand functional Δ(t) = p(t)ε₁ + (1-p(t))ε₂ and producing plots for different sign cases and initialization types.


## Import Required Libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

# Add the games module to path
sys.path.append('games')

# Import our custom classes
from exponential_weights import (
    ExponentialWeights,
    InitializationStrategies,
    GameCases,
    SimulationRunner
)

# Set up plotting
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

print("Libraries imported successfully!")
```

## Initialize Simulation Runner

The SimulationRunner class handles all aspects of running simulations and generating plots.


```python
# Initialize the simulation runner
# This will create a 'cases_output' directory for saving plots
runner = SimulationRunner(output_dir="cases_output")

print(f"Simulation runner initialized!")
print(f"Output directory: {runner.output_dir}")
print(f"Directory exists: {os.path.exists(runner.output_dir)}")
```

## Explore Predefined Game Cases

Let's examine the predefined game cases and their epsilon values.


```python
# Get all predefined cases
cases = GameCases.get_all_cases()

print("Predefined Game Cases:")
print("=" * 50)

for case_name, payoff_matrix in cases.items():
    eps1, eps2 = ExponentialWeights.compute_epsilons(payoff_matrix)
    print(f"{case_name:>6}: ε₁ = {eps1:>4.1f}, ε₂ = {eps2:>4.1f}")
    print(f"         Payoff Matrix: {payoff_matrix.tolist()}")
    print()
```

## Run All Predefined Cases

This will run simulations for all predefined cases with different initialization strategies.


```python
# Run all simulations
# This may take a few minutes depending on the number of rounds
print("Running all predefined case simulations...")
print("This will generate plots for each case and initialization strategy.")
print()

runner.run_all_simulations(eta=1.0, T=50)
```

## Analyze Results

Let's examine what files were generated in the output directory.


```python
# List all generated files
output_files = os.listdir(runner.output_dir)
output_files.sort()

print(f"Generated {len(output_files)} plot files in '{runner.output_dir}':")
print("=" * 60)

for i, filename in enumerate(output_files, 1):
    print(f"{i:2d}. {filename}")

print(f"\nTotal files: {len(output_files)}")
```
